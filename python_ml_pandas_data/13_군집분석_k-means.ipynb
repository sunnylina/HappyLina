데이터 전처리

```python
df.head(3)
```

```python
x = df.iloc[:,:]
```

```python
# 독립변수의 단위가 틀림 >> 정규화 

from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(x) # 비지도 학습 (fit>>transform)
```

```python
x
```

```python
# 비지도 학습은 데이터 분리를 하지 않음 
```

모델링

```python
from sklearn.cluster import *
```

```python
model = KMeans(n_clusters=5, n_init=10)
```

```python
# 모델 학습

model.fit(x)
```

```python
# 예측(군집)

model.labels_
```

```python
df['cluster'] = model.labels_
```

```python
df.head(3)
```

```python
# 그래프로 시각화 

df.plot(kind='scatter', x='Grocery', y='Frozen', c='cluster', 
       cmap='Set1', colorbar=False, figsize=(10,10))
df.plot(kind='scatter', x='Milk', y='Delicassen', c='cluster', 
       cmap='Set1', colorbar=True, figsize=(10,10))
plt.show()
```

```python
df.cluster
```

```python
df.cluster.value_counts()
```

```python
# 큰 값으로 구성된 클러스트 1, 3번 제외

df[(df.cluster == 1) | (df.cluster == 3)] # 제일 큰 군집 2개 
```

```python
condition = (df.cluster == 1) | (df.cluster == 3)
```

```python
df[condition]
```

```python
df[~condition] # 군집이 작은 record 들 
```

```python
# 큰 값들만 보기 (1, 3 그룹만 보기)

df[condition].plot(kind='scatter', x='Grocery', y='Frozen', c='cluster', cmap='Set1',
                  colorbar=False, figsize=(10,10))
df[condition].plot(kind='scatter', x='Milk', y='Delicassen', c='cluster', cmap='Set1',
                  colorbar=True, figsize=(10,10))
plt.show()
```

```python
s_cluster = df[~condition] # 군집이 작은 record 들 
```

```python
# 작은 그룹만 보기 

s_cluster.plot(kind='scatter', x='Grocery', y='Frozen', c='cluster', cmap='Set1',
                  colorbar=False, figsize=(10,10))
s_cluster.plot(kind='scatter', x='Milk', y='Delicassen', c='cluster', cmap='Set1',
                  colorbar=True, figsize=(10,10))
plt.show()
```
